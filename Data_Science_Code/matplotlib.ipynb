```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle = 'dashed',marker = 'D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linewidth='3')

plt.title('Rohit sharma vs virat kohli')
plt.xlabel('season')
plt.ylabel('runs scored')
```

# Bar plot

```python
children = [10,50,30,10,20]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors,children,color = 'red')
```

```python
plt.barh(colors,children,color = 'red')
```

```python
df = pd.read_csv('matplot-dataset/batsman_season_record.csv')
df.head()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='green')
```

```python
children = [10,50,30,10,20]
colors = ['red red redredredredred','blue','greengreengreen','yellow','pink']
plt.bar(colors,children,color = 'red')
plt.xticks(rotation='vertical')
```

```python

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2016']+df['2015']),label='2017')
plt.legend()
plt.show()
```

## Pie Chart

```python
data = [23,45,100,20,49]
subjects = ['eng','science','math','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
df = pd.read_csv('matplot-dataset/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','pink','black','yellow','red','orange'])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```

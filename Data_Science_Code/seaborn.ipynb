## Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
data = sns.load_dataset('tips')
data.head()
```

## scatter plot

```python
sns.relplot(x='total_bill',y='tip',kind = 'scatter',data=data)
```

## hue parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

## style parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data,style='sex')
```

## size parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data,style='sex',size='size',sizes=(15,100))
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

## Categorical data

```python
tips = sns.load_dataset('tips')
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=tips,jitter='0')
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=tips)
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

## Boxplots

```python
sns.boxplot(tips['tip'])
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=tips)
```

## Violinplots

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=tips)
```

## barplot and countplot

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=tips)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',estimator=np.mean,kind='bar',data=tips)
```

```python
sns.catplot(x='sex',kind='count',data=tips,hue='smoker')
```

## heatmaps

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
x = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='viridis')
```

## clustermap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,col_cluster=False,metric='correlation')
```

## displots

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.displot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],bins=100,kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
